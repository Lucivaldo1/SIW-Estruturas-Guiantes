```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\luciv\\Downloads\\atividade estruturas\\a.csv')
dfsiw = pd.read_csv('C:\\Users\\luciv\\OneDrive\\Desktop\\guias retangulares\\Gamma Plot 1 _ artigo _ teste.csv')
dfsiw1 = pd.read_csv('C:\\Users\\luciv\\OneDrive\\Desktop\\guias retangulares\\Gamma Plot 1 _ artigo.csv')
#df_filtrado = df[df['a [mm]'] == 4.795]
frequencia = df['Freq [GHz]']
betaArtigoMetal = df['im(Gamma(1:1)) []']
beta2ArtigoMetal = df['im(Gamma(1:2)) []']
betaArtigoSiw = dfsiw['im(Gamma(1:1)) []']
betaArtigo2Siw = dfsiw['im(Gamma(1:2)) []']
betaArtigo2Siw2 = dfsiw['im(Gamma(1:3)) []']
frequencia2= dfsiw['Freq [GHz]']
```

```python
plt.figure(figsize=(8, 5))
plt.plot(frequencia, betaArtigoMetal, label='Primeiro modo - metal', color='black',linewidth=2)
plt.plot(frequencia, beta2ArtigoMetal, label='Segundo modo - metal',linewidth=2)
plt.plot(frequencia2, betaArtigoSiw, label='Primeiro modo - siw',linestyle='dashed',linewidth=2)
plt.plot(frequencia2[60:164], betaArtigo2Siw[60:164], label='Segundo modo - siw',linestyle='dashed', color='r',linewidth=2)
plt.plot(frequencia2[0:60], betaArtigo2Siw2[0:60], color='r',linestyle='dashed',linewidth=2)
plt.xlabel('Frequência (GHz)')
plt.ylabel(r'$\beta$ (rad/m)')
plt.title(r'Constante de propagação $\beta$ (artigo)')
plt.grid(True)
plt.legend()
plt.xlim((19,100))
plt.tight_layout()
plt.savefig('constante de propagacao artigo')
plt.show()
```
