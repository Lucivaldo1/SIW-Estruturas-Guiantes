```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import c
import pandas as pd
from IPython.display import display, Math
```

```python
epsilon_r = 9.9
mi_r = 1
giga = 1e9
```

```python
def get_cutOffFrequecies(a,b,p,epslion_r,mi_r):
    '''
    :param a: largura do guia 
    :param b: altura do guia
    :param p: numero de loops
    :param epslion_r: permissividade
    :param mi_r: permeabilidade
    :return: lista com frequencias de corte
    '''
    c_ = c/(np.sqrt(epsilon_r * mi_r))
    lista = []
    for m in range(0,p,1):
        for n in range(0,p,1):
            fc = ((c_/2)*np.sqrt( (m/a)**2 + (n/b)**2))
            lista.append((m, n, fc / giga))
    return lista

'''def get_cutOffFrequecies(a,b,p):
    lista = []
    for m in range(0,p,1):
        for n in range(0,p,1):
            fc = ((c/2)*np.sqrt( (m/a)**2 + (n/b)**2))
            lista.append((m, n, fc / giga))
    return lista'''
```

## Primeiro guia retangular

```python
a = 10.668e-3#10.668e-3
b = 4.318e-3#4.318e-3
epsilon_r = 1
dados = get_cutOffFrequecies(a,b,4, epsilon_r, mi_r)
df = pd.DataFrame(dados, columns=['m','n','f_c(Ghz)'])
df = df.drop_duplicates().reset_index(drop=True)
#print(df.to_string(index=False, header=['m', 'n', 'f_c(GHz)']))
df = df.sort_values(by='f_c(Ghz)').reset_index(drop=True)
df
```
